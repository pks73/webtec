```python
# s10 b
def distancesum(x, y, n):
    total = 0  # Avoid using 'sum' as a variable name (shadows built-in)
    for i in range(n):
        for j in range(i + 1, n):
            total += abs(x[i] - x[j]) + abs(y[i] - y[j])
    return total

x = [-1, 1, 3, 2]
y = [5, 6, 5, 3]
n = len(x)
print(distancesum(x, y, n))  # Output: 22
```
